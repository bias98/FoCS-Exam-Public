### 4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for Aarhus University the value is 122-73=49). Notice that some rankings are expressed as a range

```python
cwurData = cwurData.rename(columns = {'institution' : 'university_name'})
```

```python
rankTimes = timesData.loc[timesData.groupby('university_name')['year'].idxmax()][['university_name', 'world_rank', 'year']]
rankShan = shanghaiData.loc[shanghaiData.groupby('university_name')['year'].idxmax()][['university_name', 'world_rank', 'year']]
rankCwur = cwurData.loc[cwurData.groupby('university_name')['year'].idxmax()][['university_name', 'world_rank', 'year']]
```

```python
rank_concat = pd.concat([rankTimes, rankShan, rankCwur])
#rank_concat
```

Bisogna creare una funzione che permetta di sistemare qualche valore di rank

```python
def sistema_rank(r):
    r = str(r)
    if type(r) == int:
        return r
    elif '-' in r:
        a = r.split('-')[0]
        b = r.split('-')[1]
        return ((int(a)+int(b))*0.5)
    elif '=' in r:
        a = r.strip('=')
        return (int(a))
    else:
        return (int(r))
```

```python
rank_concat['world_rank'] = rank_concat['world_rank'].apply(sistema_rank)

differenza = (rank_concat.groupby('university_name')['world_rank'].max() - rank_concat.groupby('university_name')['world_rank'].min()).reset_index()
differenza
```
